# **Perceptron Model**

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_digits
```

```python
# Load_digits is an MNIST Data set

digits = load_digits()
digits
```

From the above:

a. Pixel value is a cell value, range is between  to 0 to 255.

b. target value is 0, 1, 2.......,9, therefore its a multi-class classification.

```python
# This is my input data
digits['data']
```

```python
# This is my target variable
digits['target']
```

```python
x = digits.data
y = digits.target
```

```python
#x = x/255.0, to scall the data (so since we are using sklearn and linear, therefore no need to worry. When we use deep neural network and if using TensorFlow, Keras then we have to scale it. )
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
# Building perceptron model
from sklearn.linear_model import Perceptron
```

```python
perceptron_model = Perceptron()
perceptron_model.fit(x_train, y_train)
```

```python
print("What is the Traingin Accuracy of Perceptron Model :")
print(perceptron_model.score(x_train, y_train))

print("What is the Test Accuracy of Perceptron Model :")
print(perceptron_model.score(x_test, y_test))

predicted_train = perceptron_model.predict(x_train)
predicted_test = perceptron_model.predict(x_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Training Accuracy", accuracy_score(y_train, predicted_train))
print("Test Accuracy", accuracy_score(y_test, predicted_test))

print("classification_report Training")
print(classification_report(y_train, predicted_train))
print("classification_report Test")
print(classification_report(y_test, predicted_test))
```

From above, the Perceptron model demonstrates strong performance in both training and testing phases. **With a training accuracy of approximately 97.62% and a test accuracy of about 96.89%, the model exhibits high accuracy in classifying digit images.** Furthermore, the classification reports provide additional insights into the model's performance, showing high precision, recall, and F1-scores across various digit classes for both training and testing datasets. Overall, this Perceptron model proves to be a robust and reliable choice for digit recognition tasks.

**We do not use the above method (i.e., scikit-learn) for practical applications; it is primarily for learning purposes. However, when moving forward, especially when working with deep neural networks, we typically rely on TensorFlow and scikit-learn for our machine learning and deep learning tasks.**

**Conclusion for Perceptron Model**

The Perceptron model is a fundamental concept in machine learning, particularly useful for binary classification problems. However, for the multi-class classification task like the one presented in the MNIST dataset, a simple perceptron might not be sufficient due to its limitations in handling complex, non-linearly separable data. In such cases, more advanced models like Multilayer Perceptrons or other deep learning architectures are more suitable.

# Classification Problem

```python
import pandas as pd
dataset = pd.read_csv("/content/Churn_Modelling.csv")
dataset.head()
```

```python
dataset['Exited'].value_counts()
```

```python
dataset = dataset.iloc[:,3:]
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset = pd.get_dummies(dataset, columns=['Geography', 'Gender'], drop_first=True)
```

```python
dataset.head()
```

```python
# split the data into ind and dep variable
x = dataset.drop(['Exited'], axis=1)
y = dataset[['Exited']]
```

```python
x.shape
```

```python
x.head()
```

```python
y.head()
```

```python
# Balance the data
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_smote, y_smote = smote.fit_resample(x, y)
print(y.value_counts())
print(y_smote.value_counts())
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_smote, y_smote, test_size=0.25, random_state=101)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

This section probably covers the nuances of solving classification problems, where the goal is to categorize data into predefined classes. It's a fundamental problem type in machine learning.


# MLP

```python
model = Sequential()
model.add(Dense(32, activation='relu', kernel_regularizer = keras.regularizers.l2(0.03),
                kernel_initializer='he_normal',  input_dim=11))
model.add(Dropout(0.5))
model.add(BatchNormalization())

model.add(Dense(32, activation='relu',  kernel_regularizer = keras.regularizers.l2(0.03), kernel_initializer='he_normal'))
model.add(Dropout(0.5))
model.add(BatchNormalization())

model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0]
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'] )
```

```python
history = model.fit(x_train, y_train, batch_size=32, epochs=100, validation_data=(x_test, y_test))
```

The MLP (MultiLayer Perceptron) section likely details the structure and use of MLPs in solving various problems. MLPs are versatile and form the backbone of many neural network architectures.
